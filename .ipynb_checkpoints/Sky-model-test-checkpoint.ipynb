### Converting Alt and Az to l and m for each component. 

Once we have the $l$ and $m$ values for each component we can transform them to the Zenith value, and calculate new Azimuth and Altiture values.

```python
l_mod = np.cos(np.radians(Alt))*np.sin(np.radians(Az))# Slant Orthographic Project
m_mod = np.cos(np.radians(Alt))*np.cos(np.radians(Az))# Slant Orthographic Project

# Calculating the middle values. 
l_mid = 0.5*(np.max(l_mod) + np.min(l_mod))
m_mid = 0.5*(np.max(m_mod) + np.min(m_mod))

# Shifting by the middle values.
#l_mod_nu = l_mod - l_mid
l_mod_nu = l_mod# - l_mid
#m_mod_nu = m_mod - m_mid
m_mod_nu = m_mod# - m_mid

# Calculating the azimuth and altitude values.
Alt_nu = np.arccos(np.sqrt(l_mod_nu**2 + m_mod_nu**2)) # Alt = arccos([l^2 + m^2]^(1/2))
Az_nu = 2*np.pi - (np.arctan2(l_mod_nu,-m_mod_nu) + np.pi)  #arctan2() returns [-pi,pi] we want [0,2pi].

Alt_nu = np.degrees(Alt_nu)
Az_nu = np.degrees(Az_nu)
```
